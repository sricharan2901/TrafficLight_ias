## Storing the Images in an Excel file

```python
train_path = "/content/cinTA_v2-1/train"
test_path = "/content/cinTA_v2-1/test"
valid_path = "/content/cinTA_v2-1/valid"

dataset_paths = {
    "Train": train_path,
    "Test": test_path,
    "Validation": valid_path
}
```

```python
def get_image_info(folder, dataset_type):
    data = []
    if not os.path.exists(folder):
        print(f"‚ùå Directory not found: {folder}")
        return data
    i = 1
    for root, _, files in os.walk(folder):
        for filename in files:
            file_path = os.path.join(root, filename)
            if file_path.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tiff", ".webp")):
                img = cv2.imread(file_path)
                if img is not None:
                    height, width, channels = img.shape
                    if channels == 3:
                        channel = "RGB"
                    else:
                        channel = "Grayscale"
                    img_size = f"{height} x {width}"
                    ext = filename.split('.')[-1]
                    print(f"‚úÖ Read: {filename} | Shape: {height}x{width}x{channels} | Format: {ext}")
                    data.append([i,dataset_type, filename, img_size, channel, ext])
                    i+=1
                else:
                    print(f"‚ö†Ô∏è Skipped (Invalid image): {filename}")
    return data
```

```python
image_data = []
for dtype, path in dataset_paths.items():
    print(f"\nüìÇ Processing {dtype} dataset...")
    image_data.extend(get_image_info(path, dtype))

if not image_data:
    print("‚ùå No images were found or processed.")
```

## Checking if all images are of the same format and dimensions

```python
df = pd.DataFrame(image_data, columns=["S.No", "Dataset Type", "Filename", "Image Size", "Channel", "Extension"])

same_dim = len(df['Image Size'].unique()) == 1

if same_dim:
    print("\n‚úÖ All images have the same dimensions.")
    print(f"Size: {df['Image Size'].iloc[0]}, Channels: {df['Channel'].iloc[0]}")
else:
    print("\n‚ùå Images have different dimensions.\nDifferent Sizes: ")
    print(df['Image Size'].value_counts())
```

```python
same_extension = len(df['Extension'].unique()) == 1
if same_extension:
  print("\n‚úÖ All images have same extension")
  print(f"Extension: {df['Extension'].iloc[0]}")
else:
  print("\n‚ùå Images have different extensions.")
  print(df["Extension"].value_counts())
```
